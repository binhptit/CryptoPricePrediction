```python
binance_crypto_data_crawler = BinanceCryptoDataCrawler()
crypto_data = binance_crypto_data_crawler.load_from_file(r'dataset/btc_binance_crypto_price_data.json')

crypto_data['BTCUSDT']['4h']['data'][0]
```

```python
def make_plot_from_dataframe(d, name):
    save_path = r'/mnt/lustre/home/jason/sota_exp/CryptoPricePrediction/images/plot/' + name

    if not os.path.exists(save_path):
        os.mkdir(save_path)

    df = pd.DataFrame(data=d)
    df.index = pd.DatetimeIndex(df['Date'])

    filename = save_path + "/" + str(d['Date'][0]) + '.png'
    coin = f"BTC-{name}"
    
    mpl.plot(
        df,
        type="candle", 
        title = f"{coin} Price",  
        style="yahoo",
        savefig=filename
        )
    
```
